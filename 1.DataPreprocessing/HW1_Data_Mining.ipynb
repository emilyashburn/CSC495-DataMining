```python
# Question 3
# ----------------
# For the continuous features, what are the average, median, maximum, and minimum values? 
# What is the standard deviation?
# ----------------

# Import .tsv file
path = abspath("Video_Store.tsv")

df1 = pandas.read_csv(path, sep='\t', usecols=['Cust ID', 'Gender', 'Income', 'Age', 'Rentals', 'Avg Per Visit', 'Incidentals', 'Genre'])

print(df1)
```

```python
def Question3():

# Remove the '$' and ',' in the Income column in order to cast as float for calculations
  df1['Income'] = df1['Income'].str.replace('\$', '')
  df1['Income'] = df1['Income'].str.replace('\,', '')
  df1['Income']= df1['Income'].astype(float)

# Display all calculations 
  print("Income mean: ", df1['Income'].mean())
  print("Income median: ", df1['Income'].median())
  print("Income minimum: ", df1['Income'].min())
  print("Income maximum: ", df1['Income'].max())
  print("Income standard deviation: ",df1['Income'].std())

  print("Age mean: ", df1['Age'].mean())
  print("Age median: ", df1['Age'].median()) 
  print("Age minimum: ", df1['Age'].min())
  print("Age maximum: ", df1['Age'].max())
  print("Age standard deviation: ", df1['Age'].std())

  print("Rentals mean: ", df1['Rentals'].mean())
  print("Rentals median: ", df1['Rentals'].median())
  print("Rentals minimum: ", df1['Rentals'].min())
  print("Rentals maximum: ", df1['Rentals'].max())
  print("Rentals maximum: ", df1['Rentals'].std())

  print("Avg per visit mean: ", df1['Avg Per Visit'].mean())
  print("Avg per visit median: ", df1['Avg Per Visit'].median())
  print("Avg per visit minimum: ", df1['Avg Per Visit'].min())
  print("Avg per visit maximum: ", df1['Avg Per Visit'].max())
  print("Avg per visit standard deviation: ", df1['Avg Per Visit'].std())
  
Question3()
```
